```python
import pandas as pd
import numpy as np
```

```python
df=pd.read_csv('sales.csv')
```

```python
df
```

```python
#Fix column datatypes.
df.isnull()
```

```python
#Drop duplicates and null values
df.drop_duplicates(inplace=True)
```

```python
df.dropna(inplace=True)
```

```python
df.head()
```

```python
# Sanity check for value ranges and to check assumptions.

df = df[df['price'] >= 0]

assert (df['price'] >= 0).all(), "Price should be non-negative"
```

```python
assert (df['quantity'] > 0).all(), "Quantity should be positive"
```

```python
df.head()
```

```python
#Use regular expression and lambda function to parse data
import re
df['product_type'] = df['name'].apply(lambda x: re.search('"([^"]+)"', x).group(1))
```
